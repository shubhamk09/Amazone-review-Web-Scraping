```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link="https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
page=requests.get(link)
```

```python
page
```

```python
soup=bs(page.content,'html.parser')
print(soup.prettify())
```

```python
names=soup.find_all('span', class_='a-profile-name')
```

```python
names
```

```python
cust_name=[]
for i in names:
    cust_name.append(i.get_text())
cust_name
```

```python
#Remove Descripency
cust_name.pop(0)
cust_name.pop(0)
```

```python
title=soup.find_all('a', class_='review-title-content')
title
```

```python
review_title=[]
for i in title:
    review_title.append(i.get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
rating=soup.find_all('i', class_='review-rating')
rating
```

```python
cust_ratinge=[]
for i in rating:
    cust_ratinge.append(i.get_text())
cust_ratinge
```

```python
cust_ratinge.pop(0)
cust_ratinge.pop(0)
```

```python
review=soup.find_all('span',{"data-hook":"review-body"})
```

```python
review
```

```python
review_content=[]
for i in review:
    review_content.append(i.get_text())
review_content
```

```python
review_content[:] = [titles.rstrip('\n') for titles in review_content]
review_content
```

```python
review_content[:] = [titles.lstrip('\n') for titles in review_content]
review_content
```

## Converting to CSV

```python
import pandas as pd
```

```python
review_Data=pd.DataFrame()
```

```python
review_Data['Customer_name']=cust_name
```

```python
review_Data['Review_title']=review_title
```

```python
review_Data['Review']=review_content
```

```python
review_Data['Rating']=cust_ratinge
review_Data
```

```python
review_Data.Rating=review_Data.Rating.map(lambda str: str[0:3])
```

```python
review_Data
```

```python
rating_count=review_Data.groupby('Rating').Rating.count()
rating_count
```

```python
index=list(rating_count.index)
index
```

```python
values=list(rating_count.values)
values
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(index,values)
plt.show()
```

```python
mean_rating=rating_count.mean()
mean_rating
```
